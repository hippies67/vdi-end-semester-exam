**1. MENAMPILKAN DATA TRANSAKSI**

```python
data_transaksi = pd.read_csv("uas_datasets/transaksi2.csv", names=['date', 'id_product', 'quantity', 'id_penjualan'])
```

```python
data_transaksi
```

// Datasets Transaksi

Kode Python

```python
data_transaksi['date'] = pd.to_datetime(data_transaksi['date'])
data_transaksi['year'] = data_transaksi['date'].dt.year
data_transaksi['month'] = data_transaksi['date'].dt.month_name()
data_clean = data_transaksi

plt.title('Plot 1 - Menampilkan Data Transaksi')
plt.scatter(data_clean['date'], data_clean['quantity'])

plt.xlabel('Date')
plt.ylabel('Quantity')

plt.show()
```

**2. MENAMPILKAN GRAFIK PESANAN DARI SETIAP KANTOR**

```python
data_supply_kantor = pd.read_csv('uas_datasets/SupplyKantor.csv')
```

```python
data_supply_kantor
```

// Datasets Supply Kantor

Kode Python

```python
x = data_supply_kantor['Kota']
y = data_supply_kantor['Jumlah']
plt.figure(figsize=(10, 8))
plt.scatter(x, y)

plt.xlabel('Kota')
plt.ylabel('Jumlah')

plt.title('Plot 2 - Grafik Pesanan dari Setiap Kota')

plt.show()
```

**3. MENAMPILKAN GRAFIK KEBUTUHAN PER TAHUN SEMUA PROVINSI**

```python
data_kebutuhan_perbulan = pd.read_csv('uas_datasets/kebutuhan_per_bulan_2.csv')
```

```python
data_kebutuhan_perbulan
```

// Datasets Kebutuhan Per Bulan

Kode Python

```python
plt.figure(figsize=(12, 8))
plt.scatter(data_kebutuhan_perbulan['Propinsi'], data_kebutuhan_perbulan['tahun'], s=50, c='blue', alpha=0.7)

for i in range(len(data_kebutuhan_perbulan)):
    plt.annotate(data_kebutuhan_perbulan['kebutuhan'][i], (data_kebutuhan_perbulan['Propinsi'][i], data_kebutuhan_perbulan['tahun'][i]), ha='center', va='bottom', fontsize=8)

plt.title('Plot 3 - Grafik Kebutuhan Per Tahun Berdasarkan Semua Provinsi', fontsize=18)
plt.xlabel('Propinsi', fontsize=12)
plt.ylabel('Tahun', fontsize=12)

plt.xticks(rotation=90, fontsize=8)

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

**4. MENAMPILKAN GRAFIK KEBUTUHAN DARI PROVINSI JAWA BARAT**



```python
data = pd.DataFrame(data_kebutuhan_perbulan)

provinsi = 'JAWA BARAT'
data_provinsi = data.loc[data['Propinsi'] == provinsi]

x = data_provinsi['tahun']
y = data_provinsi['kebutuhan']

plt.title(f'Plot 4 - Grafik Kebutuhan Jawa Barat')
plt.xlabel("Tahun")
plt.ylabel("Kebutuhan")

plt.scatter(x, y, color='#D04', marker='s')

plt.xlim(min(x), max(x))

plt.show()
```

**5. MENAMPILKAN GRAFIK PESANAN DARI KOTA BANDUNG**


```python
data = pd.DataFrame(data_supply_kantor)

kota = 'Bandung'
data_kota = data.loc[data['Kota'] == kota]

x = data_kota['Kota']
y = data_kota['Jumlah']
plt.figure(figsize=(10, 8))
plt.scatter(x, y)

plt.xlabel('Kota')
plt.ylabel('Jumlah')

plt.title('Plot 5 - Grafik Pesanan dari Kota Bandung')

plt.show()
```
